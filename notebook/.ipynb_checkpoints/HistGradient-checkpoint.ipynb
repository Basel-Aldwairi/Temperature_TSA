```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.inspection import permutation_importance
```

```python
df = pd.read_csv('Final_data.csv')
```

```python
df.info()
```

```python
df.describe(include='all', percentiles=[0.99])
```

Extracting Time-based features. We convert the Datetime string to actual objects to extract cyclical patterns

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.sort_values('Datetime').reset_index(drop=True)

df['Hour'] = df['Datetime'].dt.hour
df['Month'] = df['Datetime'].dt.month
df['DayOfYear'] = df['Datetime'].dt.dayofyear
df['DayOfWeek'] = df['Datetime'].dt.dayofweek
```

```python
upper_limit = df['Wind Speed'].quantile(0.999)
df['Wind Speed'] = df['Wind Speed'].clip(upper=upper_limit)
print(f"Wind Speed clipped at 99.9th percentile: {upper_limit:.2f}")
```

```python
target = 'Temperature'
features = [col for col in df.columns if col not in ['Datetime', 'Temperature']]

X = df[features]
y = df[target]
```

```python
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
```

```python
model = HistGradientBoostingRegressor(
    max_iter=100, 
    max_depth=10, 
    learning_rate=0.1, 
    random_state=42
)
print("Training...")
model.fit(X_train, y_train)
```

model eval

```python
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"\n--- Model Performance ---")
print(f"RMSE: {rmse:.4f} degrees")
print(f"R2 Score: {r2:.4f}")
```

```python
plt.figure(figsize=(12, 6))
slice_size = 500  # Viewing a 500-hour window for clarity
plt.plot(df['Datetime'].iloc[split_idx:split_idx+slice_size], 
         y_test.iloc[:slice_size], label='Actual', alpha=0.8)
plt.plot(df['Datetime'].iloc[split_idx:split_idx+slice_size], 
         y_pred[:slice_size], label='Predicted', alpha=0.8, linestyle='--')
plt.title('Temperature Forecast: Actual vs Predicted (Test Sample)')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
perm_importance = permutation_importance(model, X_test.iloc[:1000], y_test.iloc[:1000], n_repeats=10)
sorted_idx = perm_importance.importances_mean.argsort()[-10:]

plt.figure(figsize=(10, 6))
plt.barh(np.array(features)[sorted_idx], perm_importance.importances_mean[sorted_idx])
plt.xlabel("Importance Score")
plt.title("Top 10 Features Driving Predictions")
plt.show()
```
