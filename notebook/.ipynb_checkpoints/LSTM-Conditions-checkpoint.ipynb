# Time Series LSTM for Weather Condition Classification
### Author: Basel Al-Dwairi

This notebook implements a multiclass time series classification model using an LSTM neural network to predict weather conditions (e.g., fair, foggy, sunny) based on historical weather observations.

Unlike temperature forecasting (regression), this task is formulated as a categorical sequence classification problem, where the model predicts the most likely weather condition at a future timestep.

```python
import tensorflow as tf
tf.config.set_visible_devices([], 'GPU')

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import sys
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.metrics import CategoricalAccuracy
```

## Library Imports

The following libraries are used for model development, training, and evaluation:

- TensorFlow / Keras: LSTM model construction and training
- Pandas & NumPy: data handling and numerical operations
- Scikit-learn: data scaling and evaluation metrics
- Matplotlib & Seaborn visualization of model performance
- Warnings suppressing non-critical future warnings


```python
data_path = '../data/preprocessed_data_2010_2025.csv'
```

### Dataset Loading

The preprocessed weather dataset is loaded from disk.

##### This dataset includes:
- Numerical weather measurements
- Cyclically encoded wind direction
- One-hot encoded weather condition labels

All preprocessing steps are performed in a separate pipeline to ensure modularity and reproducibility.


```python
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.sort_values('Datetime')
df = df.set_index('Datetime')
```

### Data Loading and Temporal Ordering

The preprocessed weather dataset is loaded and converted into a datetime-indexed time series.

##### All records are:
- Sorted chronologically
- Indexed by timestamp

```python
target = [t for t in df.columns if t.startswith('Condition_')]
features = df.columns
```

### Target and Feature Definition

##### Target Variables
The target variables consist of all columns representing weather conditions, identified by the prefix `Condition_*`

These columns are one-hot encoded representations of categorical weather states.

##### Input Features
All available features in the dataset are used as input, allowing the model to learn relationships between meteorological variables and weather conditions.


```python
split_date = '2022-1-1'
train_df = df.loc[:split_date]
test_df = df.loc[split_date:]
```

### Time-Based Train–Test Split

##### The dataset is split chronologically using a fixed cutoff date:

- Training data: up to 2022-01-01  
- Testing data: from 2022-01-01 onward

```python
X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()
```

```python
X_train = X_scaler.fit_transform(train_df[features])
X_test = X_scaler.transform(test_df[features])
```

```python
y_train = y_scaler.fit_transform(train_df[target])
y_test = y_scaler.transform(test_df[target])
```

### Feature Scaling

##### Min–Max normalization is applied separately to:
- Input features
- Target variable

```python
WINDOW = 24 * 3
```

```python
def make_sequence(X, y, window=24):
    X_seq, y_seq = [], []
    for i in range(len(X) - window):
        X_seq.append(X[i:i+window])
        y_seq.append(y[i+window])

    return np.array(X_seq), np.array(y_seq)
```

### Sliding Window Sequence Generation

To capture temporal dependencies, the data is transformed into fixed-length sequences using a sliding window.

##### Configuration:
- 24 hours per day
- 3 days per sequence
- Window size = 72 time steps

Each input sequence predicts the temperature at the next hour.

3 days were chosen due to technical limitations.

```python
X_train_seq, y_train_seq = make_sequence(X_train, y_train, window=WINDOW)
```

```python
X_test_seq, y_test_seq = make_sequence(X_test, y_test, window=WINDOW)
```

```python
model = Sequential([
    Input(shape=(X_test_seq.shape[1],X_test_seq.shape[2])),
    LSTM(64,activation='tanh'),
    Dense(y_test_seq.shape[1], activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss=CategoricalCrossentropy(),
    metrics=['categorical_accuracy',
            'precision',
            'recall']
)
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
model_checkpoint = ModelCheckpoint(
    '../models/lstm_conditions_best_model_full.keras',
    monitor='val_loss',
    save_best_only=True,
    save_weights_only=False
)
```

### LSTM Model Architecture

The forecasting model is built using a single LSTM layer followed by a dense output layer.

##### Architecture:
- Input layer: (window length × number of features)
- LSTM layer with 64 hidden units
- Fully connected output layer

##### Activation:
- `tanh` for the LSTM layer (standard for recurrent models)
- `softmax` for the output layer (Multiclass Classification)

##### Loss function:
- Categorical Crossentropy

##### Optimizer:
- Adam optimizer with learning rate = 0.001

```python
model.fit(X_train_seq,
          y_train_seq,
         validation_split=0.1,
         epochs=25,
         batch_size=32,
         callbacks=[model_checkpoint, early_stopping])
```

### Training Strategy

##### To improve generalization and prevent overfitting:
- Early stopping is used based on validation loss
- Model checkpointing saves the best-performing model

##### Training details:
- Epochs: up to 25
- Batch size: 32
- Validation split: 10% of training data

```python
y_pred_scaled = model.predict(X_test_seq)
```

```python
y_prec = y_scaler.inverse_transform(y_pred_scaled)
y_pred = np.argmax(y_prec,axis=1)
y_pred_report = np.round(y_prec)
y_true_rescaled = y_scaler.inverse_transform(y_test_seq)
y_true = np.argmax(y_pred_scaled, axis=1)
y_true_report = np.round(y_pred_scaled)
```

```python
precision = precision_score(y_true_report, y_pred_report,average='macro', zero_division=0)
recall = recall_score(y_true_report, y_pred_report,average='macro', zero_division=0)
f1_score_ = f1_score(y_true_report, y_pred_report,average='macro', zero_division=0)
```

```python
print(f'{'Prefomance Metrics':-^30}')
print(f'{precision = }')
print(f'{recall    = }')
print(f'{f1_score_ = }')
```

### Performance Metrics

##### The following macro-averaged metrics are reported:
- Precision

Macro averaging ensures equal importance is given to each weather condition class, regardless of class frequency.
